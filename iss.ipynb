```python
import requests
import time

data =[]

#100 pos
for pos in range(100):
    response = requests.get("https://api.wheretheiss.at/v1/satellites/25544")

    if response.status_code == 200:
        iss = response.json()
        lat = iss['latitude']
        longi = iss['longitude']
        timemark = iss['timestamp']
        data.append((lat, longi, timemark))
        print(f"{pos+1}/100: X={lat}, Y={longi}, Time={timemark}")
    else:
        print(f" can't track point {pos+1}")
        
    # every 10 secs tracker
    time.sleep(10)
```

```python
import matplotlib.pyplot as plt

lat = [pos[0] for pos in data]
longi = [pos[1] for pos in data]
timemark = [pos[2] for pos in data]

plt.figure(figsize=(16,8))
plt.plot(longi, lat, marker="o", color='pink')

starting= longi[0], lat[0]
ending= longi[-1], lat[-1]

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Time vs Longitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```
